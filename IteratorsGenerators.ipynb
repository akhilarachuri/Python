## Lambda

- lambda a function without a name;

- anonymous functions

- one line only

- single expression only

- stored in a variable

- When do you use it?

- Pass a function into another function as a parameter

- lambda variables (comma separated): expression

- lambda functions are used with other functions. Usually with map, filter, sort

```python
# Function that squares a number and returns the value
```

```python
def square(num):
    return num * num

square2 = lambda num: num * num

print(square(9))
print(square2(9))
```

```python
# lambda function for adding two numbers
```

```python
add = lambda a,b: a + b
print(add(3,10))
```

### Map
- Uses function to transform the values in a sequence
- map() applies a function to every item in an iterable.
- Returns a map object (convert to list to see results).
- Used when transforming every element in a sequence.

```python
my_list = [2, 3, 4]

# map(lambda variables (comma separated): expression)

print(list(map(lambda ele: ele*ele, my_list)))
```

- Capitalize the names in the students list

```python
students = ['john', 'jane', 'doe']
print(list(map(lambda student: student.title(), students)))
```

### Reduce
- reduce() applies a function repeatedly to reduce a sequence to a single value.
- Found in functools module.
- Used for cumulative operations like sum, product, max.

```python
from functools import reduce
print(reduce(lambda a, b: a + b, [1, 2, 3, 4])) # Sum of list
print(reduce(lambda a, b: a if a > b else b, [5, 9, 2, 7])) # Find maximum
```

### Filter

- Uses function to test the truthiness of each value in the sequence and returns a filtered list.
- function must return True or False
- Helps remove unwanted data based on a condition.

```python
nums = [1, 2, 3, 4, 5, 6]
print(list(filter(lambda x: x % 2 == 0, nums)))
```

```python
list(map(lambda x: x % 2 == 0, nums))
```

```python
print(list(filter(lambda x: x > 7, [8, 1, 20, 6])))
```

```python
names = ["tom", "ross", "alex", "john"]
print(list(filter(lambda n: 'o' in n, names)))
```

### Iterator
- An iterator is an object that gives values one at a time using next().

- It keeps track of its current position.

- It works with loops (like for) because loops keep calling next() internally.

```python
nums = [10, 20, 30]
a = iter(nums)

print(next(a))  
print(next(a))  
print(next(a))  
```

```python
print(next(a))
```

```python
s = "python"
it = iter(s)

print(next(it))  
print(next(it))  
print(next(it))  
```

### Generator
- A generator is a special type of iterator created using yield.

- It produces values lazily (one at a time, only when needed).

- Takes less memory because it doesnâ€™t store all values at once.

```python
def func(n):
    for i in range(n):
        yield i
```

```python
for n in func(5):
    print(n)
```

```python
a = func(5)
print(next(a))
print(next(a))
print(next(a))
print(next(a))
print(next(a))
```

```python
list(func(5))
```

```python
def fibonacci(n):
    a, b = 0, 1
    for _ in range(n):
        yield a
        a, b = b, a + b
        print("kjhgf")
```
