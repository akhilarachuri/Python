## Modules and Packages

- A Python module is a file with a .py extension that contains functions, variables, or classes. Modules are imported into other programs using the import statement.

- When a module is imported for the first time, Python runs its code once and keeps it in memory. If the same module is imported again, Python reuses the already-loaded module instead of running it again, so its variables behave like a single shared instance.

### Exception Handling

- A SyntaxError occurs when Python cannot understand the code structure, such as a missing quote, and is detected before execution.
- Errors that occur while the program is running are called exceptions. These are not always fatal and can be handled gracefully.
- Python uses try and except blocks to handle exceptions. Code that may cause an error is placed inside try, and the error-handling logic is written inside except.

```python
try:
    x = int(input("Enter a number: "))
    result = 10 / x
except ValueError:
    print("Error: Please enter a valid number")
except ZeroDivisionError:
    print("Error: Cannot divide by zero")
else:
    print("Result:", result)
```

- Finally (alwyas executes that block)

```python
def askint():
        try:
            val = int(input("Please enter an integer: "))
        except:
            print("Looks like you did not enter an integer!")
            
        finally:
            print("Finally, I executed!")
        print(val) 
```
