### Arithmetic operators
- Symbols like +, -, *, / are used to perform math operations.
- To add, subtract, multiply, divide, and compute values in calculations.

```python
x = 12
y = 5

print(x + y)   # addition
print(x - y)   # subtraction
print(x * y)   # multiplication
print(x / y)   # division (float result)
print(x % y)   # modulus (remainder)
print(x ** y)  # exponentiation
print(x // y)  # floor division (whole number quotient)
```

### Comparison operators
- Operators like ==, !=, <, >, <=, >= are used to compare values.
- To check conditions, often used in decision-making (if/else).

```python
x = 8
y = 10

print(x == y)   # equal to
print(x != y)   # not equal to
print(x < y)    # less than
print(x > y)    # greater than
print(x <= y)   # less than or equal to
print(x >= y)   # greater than or equal to
```

### Logical operators
- and, or, not are used to combine or reverse boolean conditions.
- To create complex conditions in code logic.

```python
a = True
b = False

print(a and b)   # AND operation
print(a or b)    # OR operation
print(not a)     # NOT operation
```

### Assignment operators
- Operators like =, +=, -=, *= are used to assign or update values.
- To store data in variables and modify those values efficiently.

```python
y = 8
y = +8    # unary plus
y = -8    # unary minus
print(y)

y += 4    # y = y + 4
print(y)
y -= 2    # y = y - 2
print(y)
y *= 5    # y = y * 5
print(y)
```

### f-string
- An f-string lets you insert variable values directly inside a string using { }. It makes string formatting easy and readable.

```python
name = "Akhila"
score = 95

msg = f"{name} scored {score} marks in the test."
print(msg)
```

- .format() lets you insert values into a string by placing {} as placeholders. The values are filled in order.

```python
city = "Hyderabad"
food = "Biryani"

print("I live in {} and I love eating {}.".format(city, food))
```
